```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cardiovascular.csv')
df
```

# Average cholestrol found in various age groups

```python
group_by_age = df.groupby('age')['cholesterol'].mean().reset_index(name='cholesterol')
group_by_age
```

```python
label = group_by_age['age']
value = group_by_age['cholesterol']

fig = plt.subplots(figsize=(19,8))

plt.plot(label, value, label = "Average cholesterol as per age" , linewidth=3.0)
plt.xlabel('ages')
plt.xticks(rotation='vertical')

plt.ylabel('cholestrol')

plt.title('Average cholesterol as per age', fontsize=20)

plt.legend()

plt.show()
```

# Correlation between systolic and diastolic blood pressure

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cardiovascular.csv')
df
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it

group_by_age_max_sys = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
group_by_age_max_sys
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it

group_by_age_max_dia = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
group_by_age_max_dia
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure

sys_label = group_by_age_max_sys['age']
sys_val = group_by_age_max_sys['systolic_blood_pressure']

fig = plt.subplots(figsize=(19,8))

plt.plot(sys_label, sys_val, label="Max systolic blood pressure", linewidth=3.0)

dia_label = group_by_age_max_dia['age']
dia_val = group_by_age_max_dia['diastolic_blood_pressure']

plt.plot(dia_label, dia_val, label="Max diastolic blood pressure", linewidth=3.0)

plt.xlabel("Age")
plt.xticks(rotation='vertical')

plt.title("Correlation between systolic and diastolic blood pressure", fontsize=20)

plt.legend()

plt.show()
```
